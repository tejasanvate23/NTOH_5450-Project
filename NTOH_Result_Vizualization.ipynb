```python
#This code is for base case Cp = 4m/s and ve(eddy viscosity) = 0.1 m2/s

# velocity.CSV file

from google.colab import files
# This will prompt you to upload a file from your local machine
uploaded = files.upload()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the velocity data from the CSV file
velocity_data = np.genfromtxt('velocity.csv', delimiter=',', skip_header=1)

# Extract i, j, u, and v components
i = velocity_data[:, 0].astype(int)
j = velocity_data[:, 1].astype(int)
u = velocity_data[:, 2]
v = velocity_data[:, 3]

# Create a 2D grid for the velocity field (30x30 grid assumed)
U = np.zeros((30, 30))  # u component of velocity
V = np.zeros((30, 30))  # v component of velocity

# Populate the velocity field arrays with u and v components
for idx in range(len(i)):
    U[i[idx], j[idx]] = u[idx]
    V[i[idx], j[idx]] = v[idx]

# Rotate the velocity field 90 degrees clockwise
U_rotated = np.rot90(U, k=-1)  # k=-1 rotates 90 degrees clockwise
V_rotated = np.rot90(V, k=-1)

# Plot the rotated velocity field using a quiver plot
plt.quiver(U_rotated, V_rotated, scale=5, color='blue')
plt.title('Rotated Velocity Field (90 degrees clockwise)')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the concentration CSV file
concentration_data = np.loadtxt('concentration.csv', delimiter=',')

# Plot the concentration field with bottom-to-top orientation
plt.imshow(concentration_data, cmap='viridis', interpolation='nearest', origin='lower')
plt.colorbar(label='Concentration')
plt.title('Concentration Field')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
